```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
import random
import shutil
import tqdm as tqdm
from pathlib import Path
```

```python
path = '../data/humedas_secas_procesadas/Recortadas/'
clases = sorted(os.listdir(path))
print(clases)
```

```python
def copy_and_rename_file(src_path, dest_path, prefix):
    try:
        # Step 1: Copy the file 
        shutil.copy(src_path, dest_path)
        
        # Step 2: Get the file's extension from the source path
        #_, file_extension = os.path.splitext(src_path)
        
        # Step 3: Combine the destination path with the new filename and extension
        #new_file_path = os.path.join(dest_path, new_filename + file_extension)
        
        # Step 4: Rename the copied file to the new name
        filename = src_path.split("/")[-1]
        new_filename = prefix + filename
        os.rename(os.path.join(dest_path, filename), os.path.join(dest_path, new_filename))
        
        print(f"File copied and renamed successfully.")
    except FileNotFoundError:
        print("Source file not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
train_dry = "../data/train_dry/"
train_wet = "../data/train_wet/"
test_dry = "../data/test_dry/"
test_wet = "../data/test_wet/"

random.seed(14)

for clase in clases:
    print("====== clase: ======", clase)
    foams = sorted(os.listdir(path + clase))
    for foam in foams:
        
        foam_types = sorted(os.listdir(path + clase + "/" + foam))
        for foam_type in foam_types:
            imgs_paths = []
            foam_path = path + clase + "/" + foam + "/" + foam_type
            cantidad = len(os.listdir(foam_path))
            print("cantidad de imagenes: ", cantidad)
            to_train = int(cantidad * 0.8)
            to_test = cantidad - to_train
            print("to train: ", to_train)
            print("to test: ", to_test)
            
            imgs = sorted(os.listdir(foam_path))
            for img in tqdm.tqdm(imgs):
               img_path = foam_path + "/" + img
               imgs_paths.append(img_path)

            random.shuffle(imgs_paths)
            random_train = random.sample(imgs_paths, to_train)
            random_test = list(set(imgs_paths) - set(random_train))

            #checking if there are some repeated images between train and test
            set1 = set(random_train)
            set2 = set(random_test)
            print("there are share elements:", bool(set1 & set2))

            if foam_type == "Humedas" or foam_type == "Humeda":     
                prefix = clase + "_wet" 
                print("========== train ==========")          
                for img in random_train:  
                    copy_and_rename_file(img, train_wet, prefix)                       

                print("========== test ==========")          
                for img in random_test:                     
                    copy_and_rename_file(img, test_wet, prefix)           
                    
            else:
                prefix = clase + "_dry"  
                print("========== train ==========")          
                for img in random_train:  
                    copy_and_rename_file(img, train_dry, prefix) 
            
                print("========== test ==========")
                for img in random_test:     
                    copy_and_rename_file(img, test_dry, prefix)                

print("done!")
```
